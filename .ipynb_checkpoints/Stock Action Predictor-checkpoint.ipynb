```python
import warnings
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import torch
import numpy as np
import random

# Define the stock and time period
ticker = "VOO"
period = "1y"  # 1 year of data

# Pull data
df = yf.download(ticker, period=period)

# Calculate Moving Averages (75-day)
df['EMA_75'] = df['Close'].ewm(span=75, adjust=False).mean()

# Calculate Price Returns (Daily Percentage Change)
df['Price_Return'] = df['Close'].pct_change() * 100

df = df[['Close', 'Volume', 'EMA_75', 'Price_Return']]

# Drop NaN values (caused by moving average calculations)
df.dropna(inplace=True)
df.reset_index(inplace=True)

print(df.shape)
print(df)

# Define Labeling Function
def label(pr):
    # Label as 1 (Buy) if Price_Return > 1%
    # Label as -1 (Sell) if Price_Return < -1%
    # Otherwise, 0 (Hold)
    if pr > 1:  # This should work fine
        return 1
    elif pr < -1:
        return -1
    else:
        return 0

print(df.columns)
# Apply the label function correctly to each row
labels = []
for index, row in df.iterrows():
    # Extract the value as a single float
    pr = row.loc['Price_Return']['']
    temp = label(pr)
    labels.append(temp)

df['Label'] = labels

print(df.head(3))  # Check the output

df = df[['Close', 'Volume', 'EMA_75', 'Price_Return', 'Label']]

# --- Standardize Data ---
X = df[['Close', 'Volume', 'EMA_75']].to_numpy()
X_mean = X.mean(axis=0)
X_std = X.std(axis=0)
X = (X - X_mean) / X_std  # Standardization

# Target variable
y = df['Label'].to_numpy()
```

```python

class log_reg:
    def __init__(self, d):
        self.w = torch.zeros(d, dtype=torch.float32)
        
    def prob(self, sample):
        x, y = sample
        x = torch.tensor([1, x[0], x[1], x[2]], dtype=torch.float32)  # Add bias term
        z = torch.matmul(self.w, x)
        prob = 1 / (1 + torch.exp(-z))
        return prob

    def grad(self, sample):
        x, y = sample
        x = torch.tensor([1, x[0], x[1], x[2]], dtype=torch.float32)
        g = (self.prob(sample) - y) * x
        return g

    def grad_update(self, sample, eta=0.00001):
        g = self.grad(sample)
        self.w -= eta * g

    def display(self):
        print(f"The logistic regression model parameters are: {self.w}")
        
    def cal_prob(self, x):
        d = len(x)
        all_one = torch.ones(d, dtype=torch.float32)
        x = list(zip(all_one, x[:, 0], x[:, 1], x[:, 2]))  # Add bias term
        x = torch.tensor(x)
        prob = 1 / (1 + torch.exp(-torch.matmul(self.w, x.T)))
        return prob

print("running")
# Standardize data
X = df[['Close', 'Volume', 'EMA_75']].to_numpy()
X_mean = X.mean(axis=0)
X_std = X.std(axis=0)
X = (X - X_mean) / X_std  # Standardization

# Target variable
y = df['Label'].to_numpy()
# ----------------- Splitting Data 80/20----------------
# Set a seed for reproducibility
np.random.seed(42)

# Shuffle indices
indices = np.arange(len(X))
np.random.shuffle(indices)

# Define the split size
split_index = int(0.8 * len(X))

# Training Data
train_indices = indices[:split_index]
X_train = X[train_indices]
y_train = y[train_indices]

# Testing Data
test_indices = indices[split_index:]
X_test = X[test_indices]
y_test = y[test_indices]

print(f"Training set size: {len(X_train)}")
print(f"Testing set size: {len(X_test)}")


# Initialize model
model = log_reg(d=4)
rows = X.shape[0]
iterations = 1000
cost_history = []

# Training
for i in range(iterations):
    num = random.randint(0, rows - 1)
    sample = (X[num], y[num])
    model.grad_update(sample, eta=0.01)

    # Cost calculation
    cost = 0
    for j in range(len(X)):
        sample = (X[j], y[j])
        prob = model.prob(sample)
        cost += -(y[j] * torch.log(prob) + (1 - y[j]) * torch.log(1 - prob))
    cost /= len(X)
    cost_history.append(cost.item())

# Plot cost over iterations
plt.plot(range(iterations), cost_history)
plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.title('Cost Function Optimization')
plt.grid(alpha=0.3)
plt.show()

model.display()
```
